```python
gender = {"Male":1, "Female":0}
polyuria = {"Yes":1, "No":0}
polydipsia = {"Yes":1, "No":0}
sudden_weight_loss = {"Yes":1, "No":0}
weakness = {"Yes":1, "No":0}
polyphagia = {"Yes":1, "No":0}
genital_thrush = {"Yes":1, "No":0}
visual_blurring = {"Yes":1, "No":0}
itching = {"Yes":1, "No":0}
irritability = {"Yes":1, "No":0}
delayed_healing = {"Yes":1, "No":0}
partial_paresis = {"Yes":1, "No":0}
muscle_stiffness = {"Yes":1, "No":0}
alopecia = {"Yes":1, "No":0}
obesity = {"Yes":1, "No":0}
class_col = {"Positive":1, "Negative":0}
```

```python
data['Gender']=data['Gender'].map(gender)
data['Polyuria']=data['Polyuria'].map(polyuria)
data['Polydipsia']=data['Polydipsia'].map(polydipsia)
data['sudden weight loss']=data['sudden weight loss'].map(sudden_weight_loss)
data['weakness']=data['weakness'].map(weakness)
data['Polyphagia']=data['Polyphagia'].map(polyphagia)
data['Genital thrush']=data['Genital thrush'].map(genital_thrush)
data['visual blurring']=data['visual blurring'].map(visual_blurring)
data['Itching']=data['Itching'].map(itching)
data['Irritability']=data['Irritability'].map(irritability)
data['delayed healing']=data['delayed healing'].map(delayed_healing)
data['partial paresis']=data['partial paresis'].map(partial_paresis)
data['muscle stiffness']=data['muscle stiffness'].map(muscle_stiffness)
data['Alopecia']=data['Alopecia'].map(alopecia)
data['Obesity']=data['Obesity'].map(obesity)
data['class']=data['class'].map(class_col)
```
